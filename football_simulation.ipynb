```python
import numpy as np
import pandas as pd
```

```python
def simulate_match(home_attack, home_defense,
                   away_attack, away_defense,
                   home_advantage):

    lambda_home = home_attack * away_defense * home_advantage
    lambda_away = away_attack * home_defense

    home_goals = np.random.poisson(lambda_home)
    away_goals = np.random.poisson(lambda_away)

    if home_goals > away_goals:
        outcome = 2
    elif home_goals < away_goals:
        outcome = 0
    else:
        outcome = 1

    return home_goals, away_goals, outcome
```

```python
data = []

for _ in range(1500):
    home_attack = np.random.uniform(0.5, 2.0)
    home_defense = np.random.uniform(0.5, 2.0)
    away_attack = np.random.uniform(0.5, 2.0)
    away_defense = np.random.uniform(0.5, 2.0)
    home_advantage = np.random.uniform(1.0, 1.5)

    home_goals, away_goals, outcome = simulate_match(
        home_attack, home_defense,
        away_attack, away_defense,
        home_advantage
    )

    data.append([
        home_attack, home_defense,
        away_attack, away_defense,
        home_advantage,
        home_goals, away_goals,
        outcome
    ])
```

```python
df = pd.DataFrame(
    data,
    columns=[
        "home_attack",
        "home_defense",
        "away_attack",
        "away_defense",
        "home_advantage",
        "home_goals",
        "away_goals",
        "match_outcome"
    ]
)

df.head()
```

```python
df.to_csv("simulated_matches.csv", index=False)
```
